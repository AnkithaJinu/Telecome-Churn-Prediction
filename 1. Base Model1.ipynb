# Label Encoder with for loop

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in col:
    le = LabelEncoder()
    df[i] = le.fit_transform(df[i])
```

```python
df.info()
```

```python
df = df.drop("index",axis=1)
```

```python
X = df.drop("Churn",axis=1)
y = df["Churn"]
```

```python
X
```

# Features scaling

```python
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(X)
```

```python
x
```

# Training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(x,y,train_size=0.75,random_state=0)
```

# Base model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model1 = RandomForestClassifier()
model1.fit(X_train,y_train)
```

# Model validation with confusion matrix and k-fold

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

y_pred = model1.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
y_test.shape,y_pred.shape
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import KFold,cross_val_score
fold = KFold(n_splits=10)
```

```python
est = [DecisionTreeClassifier(),RandomForestClassifier(),LogisticRegression()]
```

```python
for i in est:
    results = cross_val_score(i,x,y,cv= fold)
    
    print("---------------------",i)
    print("Min acc {} || Max acc {}".format(results.min(),results.max()))
    print("Std: ",results.std())
    print("Mean: ",results.mean())
```
